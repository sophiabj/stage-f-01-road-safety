```python
import pandas as pd
import sys

df1 = pd.read_csv('dftRoadSafety_Accidents_2016.csv')
df2 = pd.read_csv('Cas.csv')
df3 = pd.read_csv('MakeModel2016.csv')

pd.set_option('display.max_columns', None)
```

# Merging the Datasets

```python
df = pd.merge(df1, df2, on=['Accident_Index'])
df.head()
```

```python
df_merged = pd.merge(df, df3, on=['Accident_Index'])
df_merged
```

# Deriving "Number of Car Occupants"

```python
df_merged['Car_Passenger'].unique()
```

```python
df_merged2 = df_merged.loc[lambda df_merged: df_merged.Casualty_Class <=2, :]
```

```python
new = df_merged2[['Accident_Index', 'Number_of_Casualties', 'Casualty_Class', 'Car_Passenger', 'Bus_or_Coach_Passenger']]
new.head()
```

```python
new2 = new.loc[lambda new: new.Bus_or_Coach_Passenger ==0, :]
new2.head()
```

```python
def get_casualty(row):
    if row['Casualty_Class']==1:
        return 1
    elif row['Casualty_Class']== 2:
        return 1
    
new2['Occupants'] = new2.apply(get_casualty, axis=1)
new2.head()
```

```python
def get_car_passenger(row):
    if row['Car_Passenger']==1:
        return 1
    elif row['Car_Passenger']== 2:
        return 1
    else:
        return 1

new2['Car_Passenger'] = new2.apply(get_car_passenger, axis=1)
new2.head()
```

```python
new2['Car_Occupants'] = new2.Occupants | new2.Car_Passenger
```

```python
new2.tail()
```

```python
a = new2.groupby('Accident_Index')['Number_of_Casualties'].mean()
```

```python
b = new2.groupby('Accident_Index')['Car_Occupants'].sum()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
```

```python
c = pd.merge(a,b, on=['Accident_Index'])
```

```python
c.head(100)
```

```python
c.tail()
```

```python
c.corr()
```

```python
c['Number_of_Casualties'].unique()
```

```python
x = c['Car_Occupants'].values.reshape(-1,1)
y = c['Number_of_Casualties']
```

# Model Training and Evaluation

```python
from scipy.stats import pearsonr

stat, p = pearsonr(x, y)
stat, p
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

print(lin_reg.intercept_)
print(lin_reg.coef_)
```

```python
y_pred = lin_reg.predict(x_test)

print ('r_squared:', r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Ridge

ridgereg = Ridge(alpha=1)

ridgereg.fit(x_train, y_train)
reg_pred = ridgereg.predict(x_test)

print ('r_squared:', r2_score(y_test, reg_pred))
```

```python
from sklearn.linear_model import Lasso

lass = Lasso(alpha=0.05)

lass.fit(x_train, y_train)
lass_pred = lass.predict(x_test)

print ('r_squared:', r2_score(y_test, lass_pred))
```

```python
from sklearn.linear_model import ElasticNet

elan = ElasticNet(alpha=1)

elan.fit(x_train, y_train)
elan_pred = elan.predict(x_test)

print ('r_squared:', r2_score(y_test, elan_pred))
```
